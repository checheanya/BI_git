```python
!jellyfish count -m 31 -s 1M -C ./raw_data/SRR292678sub_S1_L001_R1_001.fastq -o ./jellyfish/SRR292678_R1.jf
```

```python
!jellyfish histo ./jellyfish/SRR292678_R1.jf -o ./jellyfish/SRR292678_R1.histo
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
k_mers_bins = []
k_mers_numbers = []
with open('./jellyfish/SRR292678_R1.histo', 'r') as file:
    for line in file:
        k_mers_bin, k_mers_number = line.split()
        k_mers_bins.append(int(k_mers_bin))
        k_mers_numbers.append(int(k_mers_number))
```

```python
plt.figure(figsize=(10, 7))
sns.set_theme()
sns.lineplot(k_mers_bins[5:201], k_mers_numbers[5:201])
plt.title("Distribution of the K-mers in the SRR292678 library")
plt.xlabel("K-mer frequency")
plt.ylabel("Counts")

plt.savefig('K_mer_freq.png')
```

```python
sum([k_mers_bins[i] * k_mers_numbers[i] for i in range(16, len(k_mers_numbers))])
```

```python
for i in range(50, 75):
    print(f'{k_mers_bins[i]} {k_mers_numbers[i]}')
# peak at 62
```

```python
sum([k_mers_bins[i] * k_mers_numbers[i] for i in range(16, len(k_mers_numbers))]) // 62
```

```python
sum([k_mers_bins[i] * k_mers_numbers[i] for i in range(16, 101)]) // 62
```

```python
sum([k_mers_bins[i] * k_mers_numbers[i] for i in range(16, 101)]) / sum([k_mers_bins[i] * k_mers_numbers[i] for i in range(16, len(k_mers_numbers))])
```
